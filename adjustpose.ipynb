```python
def pose_adjust(dnum, cand):
    posefile = 'op_results/debate'+str(dnum) + '_' + cand + '_op_fixed_new.csv'
    print(posefile)

    chunksize = 1000000
    pose = pd.read_csv(posefile, chunksize=chunksize, iterator=True)
    posedf = pd.concat(pose, ignore_index=True)
    posedf.columns = posedf.columns.str.strip()
    #posedf = posedf.replace(to_replace=0, method='ffill')
    
    newpose = posedf.copy() 
    #newpose = newpose.loc[:3]
    newpose = newpose.reset_index(drop=True)
    #newpose = newpose.iloc[:, 0:76]
    #print(newpose)

    finalpose = pd.DataFrame()
    
    #finalpose['timestamp'] = newpose['timestamp']
    #finalpose['face_id'] = 0
    finalpose['frame'] = newpose['frame']
    #cols = [0, 1, 2, 5, 15, 16, 17, 18]
    cols = range(0, 25)
    for item in cols:
        formula = (3 * item) + 1
        xstr = 'x_' + str(item)
        ystr = 'y_' + str(item)
        print(newpose[newpose.columns[formula]])
        finalpose[xstr] = newpose[newpose.columns[formula]]
        finalpose[ystr] = newpose[newpose.columns[formula+1]]

    #print(finalpose)
    finalpose.to_csv('op_results/debate' + str(dnum) + '_' + cand + '_keypoints_new.csv', index=False)

    #posemask = posedf.astype(bool)
```

```python
pose_adjust(1, 'trump')
pose_adjust(1, 'clinton')
pose_adjust(3, 'trump')
pose_adjust(3, 'clinton')
```
